```python
import pandas as pd
```

```python
df=pd.read_csv('mobile_train.csv')
```

```python
df
```

```python
df.info()
```

```python
X=df.drop(labels='price_range',axis=1)
Y=df[['price_range']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X=pd.DataFrame(ss.fit_transform(X),columns=X.columns)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from keras.utils import to_categorical
```

```python
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[:5]
```

```python
ytest_cat[:5]
```

```python
from keras.models import Sequential
from keras.layers import Input,Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
# Initialize the model
model=Sequential()

# Add input layers
model.add(Input(shape=(xtrain.shape[1],)))

# Add hidden layer 1
model.add(Dense(units=64, activation='relu'))
#add droupout layer
model.add(Dropout(0.3))
# Add hidden layer 2
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.2))
# Add output layer
model.add(Dense(units=ytrain_cat.shape[1], activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
es=EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[es])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Train Loss')
plt.plot(hist.history['val_loss'],label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('categorical crossentrophy')
plt.title('Learning Curve')
plt.show()
```

```python
model.evaluate(xtrain,ytrain_cat)
```

```python
model.evaluate(xtest,ytest_cat)
```

```python
yprob_train=model.predict(xtrain)
yprob_test=model.predict(xtest)
```
